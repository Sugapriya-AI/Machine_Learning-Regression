```python
# 1.DATA COLLECTION
dataset=pd.read_csv("SLR-data.csv")#load the dataset
```

- use to load data from a CSV (Comma-Separated Values) file into a DataFrame.
- Dataframe is table like structure with rows and columns.

```python
# view the data
dataset
```

```python
# 2. INPUT/OUTPUT SPLIT
# 2.1 Input split
independent=dataset[["Years_of_Experience"]]
```

- selecting a specific column from the dataset and storing it in a new variable called independent

```python
independent
```

```python
# 2.2 Output split 
dependent=dataset[["Salary"]]
```

- selecting a specific column from the dataset and storing it in a new variable called dependent

```python
dependent
```

```python
# 3. SPLIT THE TRAIN AND TEST SETS
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0 )
```

-  sklearn is a phython library for machine learning
-  Imports the train_test_split function from sklearn, which splits data into   training and testing sets.
- Training set: Used to train a model.
- Testing set: Used to evaluate the model’s performance
- test_size=0.3 (30% test, 70% train)
- Here dataset splitted into x_train(input training 70%),x_test(input    testing 30%),y_train(output training 70%),y_test(output testing 30%) 


```python
# 4. MODEL CREATION
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)#fit the train sets into model
```

- PROCEDURE FOR SIMPLE LINEAR REGRESSION
- Imports the LinearRegression class from sklearn’s linear_model module. 
- This is used to create a linear regression model, which predicts a continuous target variable like salary based on the experience
- "regressor = LinearRegression()"
   Creates an instance of the LinearRegression model and assigns it to the variable regressor.
   Here the model object, ready to be trained.
- "regressor.fit(x_train, y_train)"
  The model learns the relationship between x_train and y_train by finding the best-fit line (y = wx + b, where w is the weight and b is the   bias).

```python
weight=regressor.coef_
```

- In y = wx + b, w is the slope. Here, coef_ is w.

```python
weight
```

```python
bias=regressor.intercept_
```

- The intercept b shows the baseline salary

```python
bias
```

```python
# Predicts the test set
y_pred=regressor.predict(x_test)
```

- The line y_pred = regressor.predict(x_test) uses your trained LinearRegression model to make predictions on the test set (x_test) and stores the results in y_pred

```python
# 5. EVALUTION METRICS
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

- Used R squared method for validation
- R squared values exits between 0 and 1
- "r_score" value nearly to 1 --> better performance
- "r_score" value nearly to 0 --> poor performance

```python
r_score
```

- Here, r_score value is nearer to 1 So, the model is performed well

```python
# 6. SAVE THE MODEL
import pickle
filename="Finalized_model-SLR.sav"
```

- Here saving the file names as Finalized_model-SLR

```python
pickle.dump(regressor,open(filename, 'wb'))
```

- the model (including its weights, intercept, etc.) into a format that can be stored and later loaded.
- Opens a file in write-binary mode ('wb') to store the serialized model.

```python
# Run the model by giving input and check its performance
loaded_model = pickle.load(open("Finalized_model-SLR.sav", 'rb'))
input_data = pd.DataFrame({'Years_of_Experience': [13]})
result = loaded_model.predict(input_data)
```
